```python
#let's see if there is any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Data Types for our columns

print(df.dtypes)
```

```python
#change the data type of the columns

#change the data type of the columns

# Fill missing values with 0 before converting to integer
df['budget'] = df['budget'].fillna(0).astype('int64')
df['gross'] = df['gross'].fillna(0).astype('int64')
```

```python
df
```

```python
#fetch only date from the released column
df['releaseddate'] = pd.to_datetime(df['released'].str.extract('(\w+ \d+, \d+)', expand=False))
df

```

```python
#create correct year column
df['yearcorrect'] = df['releaseddate'].astype(str).str[:4]

df
```

```python
#order by gross revenue
df.sort_values(by=['gross'], inplace=False, ascending=False)
df
```

```python
# Convert 'object' columns to numerical if they represent numerical data
# For example, if 'budget' and 'gross' are object types, convert them to numeric
#df['budget'] = pd.to_numeric(df['budget'], errors='coerce')
#df['gross'] = pd.to_numeric(df['gross'], errors='coerce')

# Drop non-numerical columns before calculating correlation
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
df_numeric = df.select_dtypes(include=numerics)

# Calculate correlation on the numerical DataFrame
df_numeric.corr()
```

```python
df.corr(method='pearson')
```

```python
correlation_matrix = df_numeric.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Columns')
plt.xlabel('Numeric Columns')
plt.ylabel('Numeric Columns')
plt.show()
```

```python
df_numerized = df
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
df_numerized
```

```python
correlation_matrix = df_numerized.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Columns')
plt.xlabel('Numeric Columns')
plt.ylabel('Numeric Columns')
plt.show()
```

```python
df_numerized.corr()
```

```python
correlation_mat= df_numerized.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```
